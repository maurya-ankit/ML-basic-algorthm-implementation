# Handling missing values

> ## drop NaN values

 > - drop all rows with any NaN value

```python
df_drop = data1.dropna()
```

```python
df_drop
```

 > - drop all rows with having all NaN value

```python
df_drop = data1.dropna(how="all")
```

```python
df_drop
```

> - drop all columns with having NaN values

```python
df_drop = data1.dropna(axis='columns')
```

```python
df_drop
```

> - drop all rows with having atleas T non-NA values

```python
df_drop = data1.dropna(thresh=6)
```

```python
df_drop
```

> - look specific column/s for missing value

```python
df_drop = data1.dropna(subset=['Final','Midterm'])
```

```python
df_drop
```

## fill NaN values

> - fill all NaN with 0 (or any other number)

```python
df_fill = data1.fillna(0)
```

```python
df_fill
```

> - propogate Non-NaN values forword or backword to fill NaN

```python
df_fill = data1.fillna(method='ffill')
df_fill.head(10)
```

```python
df_fill = data1.fillna(method='bfill')
df_fill.head(10)
```

> - fill NaN values differently for different columns

```python
replace_NaNDict = {'TakeHome':0,'Final':50}
df_fill = data1.fillna(replace_NaNDict)
df_fill.head(10)
```

> - fillna with interpolation

```python
df_interpolate = data1.interpolate()
```

```python
df_interpolate.head(10)
```

default method is set to linear

can be changed to polynomial, nearest, index, time etc

> - filling missing value with mean median mode or constant with imputer function in sklearn library

```python
from sklearn.impute import SimpleImputer
```

```python
header_row = data1.columns
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
df_mean=imputer.fit_transform(data1.values)
df_mean=pd.DataFrame(df_mean,columns=header_row)
```

```python
df_mean
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy='median')
df_median=imputer.fit_transform(data1.values)
df_median[7]
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
df_mode=imputer.fit_transform(data1.values)
df_mode[7]
```
